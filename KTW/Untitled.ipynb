```python
import numpy as np
from stl import mesh

# Wczytaj istniejc siatk STL
existing_mesh = mesh.Mesh.from_file('GL00001873986_GLD0001873988_A00.stl')

# Pobierz list wierzchokw istniejcej siatki
vertices = existing_mesh.vectors.reshape(-1, 3)

# Pobierz list normalnych wektorw istniejcej siatki
normals = existing_mesh.normals.reshape(-1, 3)

# Utwrz now siatk STL o takiej samej liczbie wierzchokw, normalnych i cian
closed_mesh = mesh.Mesh(np.zeros(existing_mesh.vectors.shape[0], dtype=mesh.Mesh.dtype))

# Przypisz wierzchoki, normalne i ciany z istniejcej siatki do zamknitej siatki
closed_mesh.vectors = existing_mesh.vectors
closed_mesh.normals = existing_mesh.normals
closed_mesh.v0 = existing_mesh.v0
closed_mesh.v1 = existing_mesh.v1
closed_mesh.v2 = existing_mesh.v2

# Pobierz indeksy niepowtarzajcych si wierzchokw, aby znale otwarte krawdzie
unique_vertex_indices, inverse_indices = np.unique(vertices, axis=0, return_inverse=True)

# Dla kadej krawdzi, ktra pojawia si tylko raz, dodaj cian, aby j zamkn
edge_counts = np.bincount(inverse_indices)
open_edges = unique_vertex_indices[edge_counts == 1]

for edge in open_edges:
    v1, v2 = edge
    new_face = np.array([[v1, v2, v1 + v2]], dtype=mesh.Mesh.dtype)
    closed_mesh.vectors = np.concatenate((closed_mesh.vectors, new_face))

# Zapisz zamknit siatk do pliku STL
closed_mesh.save('closed_mesh.stl')
```

```python
# Wykorzystaj funkcj get_mass_properties() do uzyskania informacji o objtoci
volume, _, _, _ = closed_mesh.get_mass_properties()

# Wywietl objto zamknitej siatki
print("Objto zamknitej siatki: ", volume)
```
